```python
# imports 
from tensorflow import keras 
import tensorflow as tf
import numpy as np 
import gradio as gr
from tensorflow.keras.models import load_model

"""
Please ensure to upload the models to the active session.
This is done my going to Files->Upload Session to Storage-> select model
"""
```

```python
def predict_model_shep_city_council(input):
  # load model
  model = keras.models.load_model('preprocessed_augmented_convnet_6_classes.keras')

  # define labels
  labels = ['battery','biological','cardboard','clothes','plastic bags','polystyrene foam']

  # reshape input into (1,299,299,3)
  input = input.reshape((-1, 299, 299, 3))

  # predict image
  prediction = model.predict(input).flatten()
  
  # print prediction
  print(prediction)

  # create dictionary for gardio Label
  confidences = {labels[i]: float(prediction[i]) for i in range(6)}

  # print confidences  
  print(confidences)

  # return confidence dictionary 
  return confidences
```

```python
def predict_model_melb_city_council(input):
  # load model
  model = load_model('model_tfl.h5')

  # define labels
  labels = ['nonRecyclable',  'organic', 'recyclable']

  # reshape input into (1,299,299,3) where 1 represents a batch size of 10 
  input = input.reshape((-1, 224, 224, 3))

  # preprocess image 
  input = tf.keras.applications.mobilenet_v2.preprocess_input(input)

  # predict image
  prediction = model.predict(input).flatten()
  
  # print image
  print(prediction)

  # create dictionary for gardio Label
  confidences = {labels[i]: float(prediction[i]) for i in range(3)}

  # print confidences  
  print(confidences)

  # return confidence dictionary 
  return confidences
```

```python
# using Gradio Blocks to develop UI 
with gr.Blocks() as demo:
    gr.Markdown("SIT744 ~ Deep Learning High Distinction Task. Created by Mark Sturtz & Geraldine Jennifer Dessa")  

    # Tab for Melbourne City Council.
    with gr.Tab("Melbourne City Council"):
       with gr.Row():
           # define input as Image
           m_img_input = gr.Image(shape=(224, 224))
           # define output as Label
           m_img_output = gr.Label()
       # define button    
       m_img_button = gr.Button()

    with gr.Tab("Greater Shepparton City Council"):
        with gr.Row():
          # define input as Image
          s_img_input = gr.Image(shape=(299,299))
          # define output as Label
          s_img_output = gr.Label()
        # define button   
        s_img_button = gr.Button()
    
    # event caller that calls predict_model_melb_city_council when the button is clicked 
    m_img_button.click(predict_model_melb_city_council, inputs=m_img_input, outputs=m_img_output)
    
    # event caller that calls predict_model_shep_city_council when the button is clicked 
    s_img_button.click(predict_model_shep_city_council, inputs=s_img_input, outputs=s_img_output)

# launches gradio application 
demo.launch(debug=True)
```
